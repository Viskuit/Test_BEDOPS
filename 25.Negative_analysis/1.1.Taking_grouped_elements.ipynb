```python
%%skip
# =============================================================================
# main function
# =============================================================================
def data_save_load(option):
    """
    This function is used to save or load data for the jupyter notebook
    """
    path_folder = "ipynb_db"  # Folder to save variables
    os.makedirs(path_folder, exist_ok=True)  # Create folder if not exist
    notebook_name = os.path.basename(os.path.abspath(''))
    path_file = os.path.join(path_folder, f"{notebook_name}.variables.pkl") # Path to save the variables

    if option == "save":
        with open(path_file, "wb") as pickle_file:
            dict_variables = {
            }
            pickle.dump(dict_variables, pickle_file)
    elif option == "load":
        with open(path_file, "rb") as pickle_file:
            variables = pickle.load(pickle_file)
        # Now load the variables
        for pickle_key, pickle_value in variables.items():
            print(f"* Loading variable: {pickle_key}")
            globals()[pickle_key] = pickle_value
# =============================================================================
# Call the function
# =============================================================================
data_save_load(option="load")
```

# Modify data

## Relationship data

```python
# From "rel_df" remove all rows whose second column is "NaN"
rel_df = rel_df.dropna(subset=[1])
print(rel_df.shape)
rel_df
```

```python
rel_df = rel_df.copy()
for _, row in rel_df.iterrows():
    for elem in row.dropna():
        new_elem = elem.split("|")[1]
        row[row == elem] = new_elem
rel_df
```

```python
# How many elements are there?
counter = 0
for _, row in rel_df.iterrows():
    counter += len(row.dropna())
print(f'There are {counter} elements in total.')
```

## LRE data

```python
# In "lre_df" remove all rows where the string in "name" doesn't end in a "letter"
lre_df = lre_df[lre_df['name'].str.match(r'.*[A-Za-z]$')]
print(lre_df.shape)
lre_df
```

## Checking problem

```python
# Let's check why rel_df and lre_df doesn't have the same length
names_rel = [elem for _, row in rel_df.iterrows() for elem in row.dropna()]
print(len(names_rel))
print(type(names_rel))
```

```python
names_lre = lre_df['name'].tolist()
print(len(names_lre))
print(type(names_lre))
```

```python
rejected_errors = []
for elem in names_lre:
    if elem not in names_rel:
        print(f"Element {elem} is not in lre_df")
        rejected_errors.append(elem)
```

Ok, it was because these elements were here, which were an error in BLASTn search. Where BLASTn couldn't find them

```python
# Remove them
lre_df = lre_df[~lre_df['name'].isin(rejected_errors)]
print(lre_df.shape)
```

```python
# Remove unused variables
del(names_rel, names_lre, rejected_errors)
```

# Renaming

## rel_df

```python
# Let's rename the data in "rel_df"
lre_name = 'lre'  # From 'leishmania repetitive element'
lre_group_counter = 0
for _, row in rel_df.iterrows():
    lre_group_counter += 1  # So the first group starts in 0
    lre_group_member_counter = 0
    for elem in row.dropna():
        lre_group_member_counter += 1  # So the first group member starts in 1
        old_name = elem
        new_name = f"{lre_name}_{lre_group_counter}.{lre_group_member_counter}"
        row[row==elem] = f"{old_name}|{new_name}"

# Let's check the results
rel_df
```

## lre_df

```python
# Let's take a list from the "rel_df"
rel_names_list = [elem for _,row in rel_df.iterrows() for elem in row.dropna()]

# Let's create a dict with each element. Split by "|" and the left will be the key and the right the value:
rel_names_dict = {}
for elem in rel_names_list:
    key, value = elem.split("|")
    rel_names_dict[key] = value
```

```python
# Now let's change the data in "lre_df" column "name". For each name found in "rel_names_dict" keys, replace it for its value
lre_df.loc[:, 'name'] = lre_df['name'].map(rel_names_dict)
```

```python
# Check for NaN values in the 'name' column
print(f'Number of NaN values: {lre_df.loc[:, 'name'].isna().sum()}')
lre_df
```

# GFF creation

```python
# Columns needed for GFF file
COLUMN_NAMES = ['seqname', 'source', 'feature', 'start', 'end', 'score', 'strand', 'frame', 'attribute']

def create_gff_line(gff_row, gff_attribute_col, gff_source, gff_feature=None, gff_strand=None):
    """
    :param gff_row: Dictionary containing data for creating GFF line; keys should include 'sseqid', 'sstart', 'send', 'sstrand'
    :param gff_attribute_col: Key in gff_row to be used for GFF attribute column; if set to '.', defaults to '.'
    :param gff_source: Optional source field for GFF line; defaults to None
    :param gff_feature: Optional feature field for GFF line; defaults to None
    :param gff_strand: Optional strand field for GFF line; if not provided, defaults to '.'
    :return: Dictionary representing a GFF line with provided information
    """
    row_gff = {
        'seqname': gff_row['sseqid'],
        'source': gff_source,
        'feature': gff_feature,
        'start': gff_row['sstart'],
        'end': gff_row['send'],
        'score': '.',
        'strand': gff_row['sstrand'] if gff_strand else '.',  # Simplified condition for gff_strand
        'frame': '.',
        'attribute': f'ID={gff_row[gff_attribute_col]}' if gff_attribute_col != '.' else '.',  # Simplified condition for gff_attribute_col
    }
    
    return row_gff


def create_gff(data_frame, gff_attribute_col, gff_source, gff_feature=None, gff_strand=None):
    """
    :param data_frame: Pandas DataFrame containing the data to be converted to GFF format.
    :param gff_attribute_col: The column in the DataFrame that contains the attributes to be included in the GFF output.
    :param gff_source: The source of the GFF entries, can be None or a specific string.
    :param gff_feature: The feature type for the GFF entries, can be None or a specific string.
    :param gff_strand: The strand information for the GFF entries, can be None or a specific string.
    :return: A new Pandas DataFrame containing the formatted GFF entries.
    """
    pre_gff = [
        create_gff_line(gff_row, gff_attribute_col=gff_attribute_col, gff_source=gff_source)
        for _, gff_row in data_frame.iterrows()
    ]
    return pd.DataFrame(pre_gff, columns=COLUMN_NAMES)

# Let's save the GFF data about 'lre_gff'.
lre_gff = create_gff(lre_df, gff_source='CBM', gff_attribute_col='name')
lre_gff.head()
```

```python
# Let's save the data
lre_gff.to_csv("./data_1.1/lre.gff", header=None, index=False, sep='\t')
```
