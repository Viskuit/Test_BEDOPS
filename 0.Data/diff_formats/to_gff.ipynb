```python
# Import needed modules
import numpy as np
import pandas as pd
```

```python
main_data = pd.read_csv("../To_analyze/Results_software/SIDER2_chr32/RUNS/run_20.csv", sep=",", header=None)
main_data.head()
```

```python
main_data = main_data.dropna(axis=1, how="all")
main_data.columns = range(main_data.columns.size)
main_data.head()
```

IN the GFF files, the "end" coordinates is always greater than the "start" coordinate, so I need to change it.

```python
# Let's load data from "plus" and "minus"
main_data_plus = main_data[main_data[4] == "plus"].copy()
main_data_plus[4] = main_data_plus[4].replace("plus", "+", regex=True)
main_data_minus = main_data[main_data[4] == "minus"].copy()
```

```python
# See data from the minus strand and how 2 > 3 and it should be the opposite.
main_data_minus.head()
```

```python
# Change it so 3 > 2
main_data_minus_corrected = main_data_minus[[0, 1, 3, 2, 4, 5]]
main_data_minus_corrected.columns = range(main_data_minus_corrected.columns.size)
main_data_minus_corrected[4] = main_data_minus_corrected[4].replace("minus", "-", regex=True)  # Change "minus" to "-" for the GFF format
main_data_minus_corrected.head()
```

```python
main_data_to_GFF = pd.concat([main_data_plus, main_data_minus_corrected])
main_data_to_GFF.head()
```

```python
GFF_data = pd.DataFrame({
    'seqname': main_data_to_GFF[0],  # assuming this is a list or series of sequence names
    'source': "CBM-302",
    'feature': "SIDER",  # Ensure this is the correct feature name
    'start': main_data_to_GFF[2],
    'end': main_data_to_GFF[3],
    'score': ".",  # Placeholder if no score data
    'strand': main_data_to_GFF[4],
    'frame': ".", # Placeholder if no frame data
    'attribute': "."
})
GFF_data.head()
```

```python
GFF_data.to_csv("sider_project.gff", sep="\t", header=False, index=False)
```
