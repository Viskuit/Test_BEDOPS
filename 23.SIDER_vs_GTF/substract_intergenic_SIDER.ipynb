```python
import pandas as pd
```

```python
sider_df = pd.read_csv('./data/sider_relationship.csv', sep=",", header=0)
print(sider_df.shape)
print(sider_df.dtypes)
sider_df.head()
```

```python
subset_df = sider_df.loc[
    sider_df['gene_name'] == 'intergenic',
    ['sider_name', 'sider_chrom', 'sider_start', 'sider_end']
]
print(subset_df.shape)
print(subset_df.dtypes)
subset_df.head()
```

```python
# Create new data frame with what is needed
new_df = pd.DataFrame({
    'seqname' : subset_df['sider_chrom'],
    'source' : 'CBM',
    'feature' : 'SIDER_INTERGENIC',
    'start' : subset_df['sider_start'],
    'end' : subset_df['sider_end'],
    'score' : '.',
    'strand' : '.',
    'frame' : '.',
    'attributes' : subset_df["sider_name"]
})

# Let's add "ID=" to the 'attributes' column:
new_df['attributes'] = 'ID=' + new_df['attributes']

# Let's check the data frame
print(new_df.shape)
print(new_df.dtypes)
new_df.head()
```

```python
# save the subset as a tab separated .gff file
new_df.to_csv('./data/sider_intergenic.gff', sep='\t', index=False, header=False)
```
