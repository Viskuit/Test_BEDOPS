```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data preparation
data = {
    'Category': ['Chr32 SIDERs', 'Chr32 SIDERs', 'LmSIDEs', 'LmSIDEs'],
    'Units': [2169, 15, 2136, 20],
    'Label': ['shared', 'unique', 'shared', 'unique']
}

df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x='Category', y='Units', hue='Label', data=df)

# Adding title and labels
plt.title('Distribution of Units in Chr32 SIDERs and LmSIDEs')
plt.xlabel('Category')
plt.ylabel('Units')

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))  # Prepare fig size
bar_plot = sns.barplot(x='Category', y='Units', hue='Label', data=df)  # prepare barplot labels

# Adding values to the bars
for p in bar_plot.patches:  # Iterate for every of the 4 bars
    height = p.get_height()  # Get the height of each bar
    bar_plot.annotate(format(height, '.0f'),  # Annotate the bar with the height value
                      (p.get_x() + p.get_width() / 2., height),  # Calculate the center of the bar coor.
                      ha = 'center', va = 'center',  # Set the alignment of the text (horizontal and vertical)
                      xytext = (0, 9),  # Set the position of the text
                      textcoords = 'offset points')

# Adding title and labels
plt.title('Distribution of Units in Chr32 SIDERs and LmSIDEs')
plt.xlabel('Category')
plt.ylabel('Units')

# Display the plot
plt.show()
```
