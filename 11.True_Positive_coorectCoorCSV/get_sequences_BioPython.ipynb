```python
# Load genome
genome_path = "./data/dict/TriTrypDB-67_LinfantumJPCM5_Genome.fasta"
genome = SeqIO.to_dict(SeqIO.parse(genome_path, "fasta"))
# Load chromosome sequences
chromosome_files = os.listdir("./data/SIDERs")
chromosome_files = sorted(chromosome_files, key=lambda x: int(re.search(r"Chr(\d+)", x).group(1)))
os.chdir("/home/viskuit/Documents/Work_CBMSO/Testing_Leishmania_project/11.True_Positive_coorectCoorCSV/data/SIDERs")
chromosomes = {f"LinJ.{str(i+1).zfill(2)}": list(SeqIO.parse(chromosome_files[i], "fasta")) for i in range(36)}
```

```python
from Bio.Seq import Seq

def find_sequence_coordinates(reference, sequence):
    seq_str = str(sequence.seq)
    ref_str = str(reference.seq)
    start = ref_str.find(seq_str)
    if start != -1:
        start += 1  # Since Python is 0-based
        end = start + len(seq_str) - 1
        return start, end
    return None

coordinates = {}

for chrom_name, sequences in chromosomes.items():
    reference = genome[chrom_name]
    coordinates[chrom_name] = []
    for seq_record in sequences:
        coord = find_sequence_coordinates(reference, seq_record)
        if coord:
            coordinates[chrom_name].append((seq_record.id, coord[0], coord[1]))
        else:
            coordinates[chrom_name].append((seq_record.id, None, None))

# Print coordinates
for chrom_name, coords in coordinates.items():
    print(f"Chromosome {chrom_name}:")
    for seq_id, start, end in coords:
        print(f"{seq_id}: {start}-{end}")
```
