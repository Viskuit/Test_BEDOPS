```python
import pandas as pd
import numpy as np
```

```python
path_file = "./negative_noCDS_relationship/negative_noCDS_relationship_named.csv"
df = pd.read_csv(path_file, sep=",", header=None)
print(df.shape)
print(df.dtypes)
df.head()
```

```python
# Let's make a list of all the values in the pd dataframe without NaN values
values = df.values.flatten().tolist()
values = [value for value in values if pd.notnull(value)]
values = sorted(values, key=lambda x: int(x.split("_")[1]))
print(len(values))
print(values)
```

Why the hell the length of the elements is 680 when the real negative database is 682?!

```python
# Let's get a list with the numbers in "Seq_{x}_LinJ". Only the "x"
numbers = [int(value.split("_")[1]) for value in values]
print(numbers)
```

```python
# Check which IDs doesn't appear
for i in range(680+1):
    if i not in numbers:
        print(i)
```

```python
# Load the original database corrected
path_db = "./negative_noCDS_relationship/negative_database_nomatch_corrected.csv"
og_db = pd.read_csv(path_db, sep=",", header=0)
print(og_db.shape)
print(og_db.dtypes)
og_db.head()
```

```python
# Let's add a column with np.nan values
og_db["name"] = np.nan
og_db.head()
```

We can see how there are 682 elements!!

```python
og_db['name'] = np.nan  # This line initializes the column with NaN, which is of dtype float64

# Change the dtype of 'name' column to object (which is used for string data)
og_db['name'] = og_db['name'].astype('object')

# Now proceed with your loop
for _, value in enumerate(values):
    new_name = value.split("|")[1]
    index = int(value.split("_")[1])
    og_db.loc[index, 'name'] = new_name

og_db.head()
```

```python
# We now there are two missing values: index 319 y 505
og_db.loc[
    (og_db["name"].isnull()),
    "name" # Column to be updated
] = "rejected_noCDS_BLAST_ERROR"
```
