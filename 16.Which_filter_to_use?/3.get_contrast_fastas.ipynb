## Get neg_data that didn't match LmSIDER2A|B

```python
# Needed modules
import pandas as pd
from Bio import SeqIO
```

```python
# Import filtered negative data from consensus + LmSIDER2A|B data
neg_data = pd.read_csv("./data/consensus+LmSIDER2A+B/negatives_testing_elements.csv", sep=",", header=0)
print(neg_data.shape)
print(neg_data.dtypes)
neg_data.head()
```

```python
# Import matches from LmSIDER2A|B data with LmSIDER2A|B hallmarks
hallmark_matches = pd.read_csv("./data/consensus+LmSIDER2A+B/negatives_testing_elements_matches_with_LmSIDER2A+B.csv", sep=",", header=0)
print(hallmark_matches.shape)
print(hallmark_matches.dtypes)
hallmark_matches.head()
```

```python
# Filter matches with evalue <= 10^-3
hallmark_matches_filtered = hallmark_matches[hallmark_matches["evalue"] <= 10**-3].sort_values(by=["evalue"]).copy()  # evalue <= 10^-3
hallmark_matches_leftovers = hallmark_matches[~hallmark_matches.index.isin(hallmark_matches_filtered.index)].copy()  # leftovers
```

```python
hallmark_matches_filtered.head()
```

```python
hallmark_matches_leftovers.head()
```

```python
# Get index from sseqid, set as int and sort by index
hallmark_matches_filtered["index"] = hallmark_matches_filtered["sseqid"].str.extract(r'_(\d+)_').astype(int)
hallmark_matches_filtered.sort_values(by="index", ascending=True, inplace=True)

hallmark_matches_leftovers["index"] = hallmark_matches_leftovers["sseqid"].str.extract(r'_(\d+)_').astype(int)
hallmark_matches_leftovers.sort_values(by="index", ascending=True, inplace=True)
```

```python
# To contrast need the index from hallmark_matches_filtered
index_filtered = hallmark_matches_filtered["index"].unique().tolist()
print(len(index_filtered))
print(type(index_filtered))
print(index_filtered)
```

```python
# Now we are interested in the hallmark_matches_leftovers index
index_leftovers = hallmark_matches_leftovers["index"].unique().tolist()
print(len(index_leftovers))
print(type(index_leftovers))
print(index_leftovers)
```

```python
# Remove numbers in index_leftovers that are in index
index_leftovers = [x for x in index_leftovers if x not in index_filtered]
print(len(index_leftovers))
print(type(index_leftovers))
print(index_leftovers)
```

## Get the data from negatives_testing_elements.csv that is not in the hallmark_matches

```python
# add index column to hallmark_matches
hallmark_matches["index"] = hallmark_matches["sseqid"].str.extract(r'_(\d+)_').astype(int)
hallmark_matches.sort_values(by="index", ascending=True, inplace=True)
print(hallmark_matches.shape)
print(hallmark_matches.dtypes)
hallmark_matches.head() 
```

```python
# Get the index in a list
index_matches = hallmark_matches["index"].unique().tolist()
print(len(index_matches))
print(type(index_matches))
print(index_matches)
```

```python
# Define function to extract fasta sequences
def fasta_extractor(pathfile, outfile, extract_list):
    with open(outfile, "w") as out_file:
        # Remember "enumerate" starts in "1"
        for count, fasta in enumerate(SeqIO.parse(open(pathfile), "fasta"), start=0):  # from Bio import SeqIO
            # name, sequence = fasta.id, str(fasta.seq)
            if count in extract_list:
                SeqIO.write(fasta, out_file, "fasta")

def fasta_extractor_exclusive(pathfile, outfile, extract_list):
    with open(outfile, "w") as out_file:
        # Remember "enumerate" starts in "1"
        for count, fasta in enumerate(SeqIO.parse(open(pathfile), "fasta"), start=0):  # from Bio import SeqIO
            # name, sequence = fasta.id, str(fasta.seq)
            if count not in extract_list:
                SeqIO.write(fasta, out_file, "fasta")
```

## Get fasta

```python
# Get fasta from the elements in the negatives_testing_elements.fasta that are did not match with LmSIDER2A|B hallmarks in a BLASTn
fasta_extractor_exclusive(pathfile="./data/consensus+LmSIDER2A+B/negatives_testing_elements.fasta",
                          outfile="./data/consensus+LmSIDER2A+B/neg_dat_filter_decider/contrast/neg_data_no_LmSIDER2A+B/neg_data_no_LmSIDER2A+B.fasta",
                          extract_list=index_matches)  # Will get the fasta sequences that are not in index_matches
```

```python
# Get fasta from the elements in the negatives_testing_elements.fasta that are did not match with LmSIDER2A|B hallmarks in a BLASTn and have evalue <= 10^-3
fasta_extractor(pathfile="./data/consensus+LmSIDER2A+B/negatives_testing_elements.fasta",
                outfile="./data/consensus+LmSIDER2A+B/neg_dat_filter_decider/contrast/neg_data_no_LmSIDER2A+B_and_evalue/neg_data_no_LmSIDER2A+B_and_evalue.fasta",
                extract_list=index_leftovers)  # Will get the fasta sequences that are in index_filtered
```
